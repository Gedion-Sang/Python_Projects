```python
# Body Mass Index
# BMI = weight/height^2
def calc_bmi():
    height = df['Height']
    weight = df['Weight']
    bmi = []
    for h, w in zip(height, weight):
        bmi = w/(h**2)
    return bmi
```

```python
calc_bmi()
```

```python
BMI = calc_bmi()
```

```python
df['BMI'] = BMI
```

```python
# to get bmi for the whole data in weight-height.csv
def get_bmi():
    height = df['Height']
    weight = df['Weight']
    bmi = []
    for h, w in zip(height, weight):
        bmi.append(w/(h**2))
    return bmi
```
